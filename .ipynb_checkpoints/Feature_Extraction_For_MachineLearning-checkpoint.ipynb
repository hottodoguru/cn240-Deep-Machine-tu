```python
import cv2
import imutils
import matplotlib.pyplot as plt
```

# cup

```python
img = 'C:\\Users\\sutic\\Desktop\\AI_augment_image_glaucoma\\0Glaucoma_jpg\\366.jpg'
img = cv2.imread(img)
b,g,r = cv2.split(img)
plt.imshow(img)
```

```python
ret,mask = cv2.threshold(g,198,255,cv2.THRESH_BINARY)
k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(200,200))
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, k)
cnts = cv2.findContours(closing.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)

center = None
if len(cnts) > 0 :
    c = max(cnts, key=cv2.contourArea)
    ((x, y), radius) = cv2.minEnclosingCircle(c)
    if radius > 0 :
        print(radius*2)
```

```python
#plot cup mask
plt.figure(figsize=(20,20))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("image")
plt.subplot(2,3,2)
plt.imshow(mask)
plt.title("mask")
plt.subplot(2,3,3)
plt.imshow(closing)
plt.title("closing")
plt.show()
```

# disc

```python
ret,mask = cv2.threshold(r,254,255,cv2.THRESH_BINARY)
k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(200,200))
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, k)
cnts = cv2.findContours(closing.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)

center = None
if len(cnts) > 0 :
    c = max(cnts, key=cv2.contourArea)
    ((x, y), radius) = cv2.minEnclosingCircle(c)
    if radius > 0 :
        print(radius*2)
plt.imshow(mask)
```

```python
#plot disc mask
plt.figure(figsize=(20,20))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("image")
plt.subplot(2,3,2)
plt.imshow(mask)
plt.title("mask")
plt.subplot(2,3,3)
plt.imshow(closing)
plt.title("closing")
plt.show()
```

# exudate

```python
ret,mask = cv2.threshold(r+b,230,255,cv2.THRESH_BINARY)
k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(200,200))
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, k)
cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
center = None
if len(cnts) > 0 :
    c = max(cnts, key=cv2.contourArea)
    ((x, y), radius) = cv2.minEnclosingCircle(c)
    cv2.drawContours(img, cnts, -1,(0, 55, 255), 2)
    print(cv2.contourArea(c))
```

```python
#plot exudate mask
plt.figure(figsize=(20,20))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("image")
plt.subplot(2,3,2)
plt.imshow(mask)
plt.title("mask")
plt.subplot(2,3,3)
plt.imshow(closing)
plt.title("closing")
plt.show()
```
